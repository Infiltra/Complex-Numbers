Complex numbers are often called as imaginary numbers. But to establish what is imaginary we first need to establish what is a real quantity. Consider next cell where we give a number line and represnt a real number for establishing the concept of real quantity.

```python
x_real = np.linspace(-5, 5, 5)
y_real = np.zeros(len(x_real))
print(x_real)
plt.figure(figsize=(8, 8))
plt.hlines(0, -5, 5)
plt.vlines(0, -1, 1)
plt.scatter(x_real, y_real)
plt.grid()
plt.xlabel("Real Axis")
plt.ylabel("Imaginary Axis")
```

Now from the above figure we have an simple numpy array that has values along the x-axis alone. Due to tradition and representation let us name the x-axis as real axis in communication systems as $I$ or the In-phase component and y-axis as imaginary in communication system as $Q$ or the quadrature component. 

Now given given any number and its square-root can be sufficiently and easily represented in the above given number along it's x-axis, like say $\sqrt(4)$=2.

Now what it is we can do if we have a square root of a negative number say, $\sqrt(-1)$ or $\sqrt(-4)$, how can this be represented as an equivalent form of real numbers?. The real question is, "Is it possible to represent such numbers in our axis?". The short answer is NO. Hence the notion of complex numbers or imaginary numbers.

Now this satisfies an imaginary part notion of the number. But what does that $j$ mean?

For that let us consider the most common representation of a complex number called the Rectangular Form, which is $z = x + jy$. Where, $x$ is called the real part and $y$ is called the imaginary part and the $j$ represents the imaginary unit of the number that satisfies the above expression such that $j^2 = -1$, as no real number can actually satisfy that equation.

In the next cell let us generate some random numbers and convert them into complex numbers and plot them in an above diagram which is called as an "Argand" or "Complex" plane diagram. In out aricle and worls we'll be calling it Complex Plane diagram.

```python
x_real = np.random.randint(-5, 5, size=(5))
print(x_real)
y_imag = np.random.randint(-5, 5, size=(5))
print(y_imag)
z = x_real + 1j*y_imag
print(z)
plt.figure(figsize=(8, 8))
plt.hlines(0, -5, 5)
plt.vlines(0, -5, 5)
plt.scatter(x_real, y_imag, label="x, y")
plt.grid()
plt.title('Real and Imaginary Plot')
plt.xlabel("Real Axis")
plt.ylabel("Imaginary Axis")
plt.legend(loc="upper left")
```

From the above plot let us consider only one set of values from all of the points and explain it here. 

```python
x_values = [0, x_real[4]]
y_values = [0, y_imag[4]]
plt.figure(figsize = (8, 8))
plt.scatter(x_real[4], y_imag[4], label="x, y")
plt.plot(x_values, y_values)
plt.hlines(0, -5, 5)
plt.vlines(0, -5, 5)
plt.grid()
plt.title('Real and Imaginary Plot')
plt.xlabel("Real Axis")
plt.ylabel("Imaginary Axis")
plt.legend(loc="upper left")
```

Now from the above diagram we can make another represntation of complex number called the magnitude, angle form of the number. This is used for descriptive purposes of a complex number. This is given by the representation $X\angle \phi$.

Now for the explanation part $X$ is the magnitude or the real part of the equation that is given in the graph and $\phi$ is the angle component of the
complex number such that it is the direction of the complex number, such that $\phi = \tan ^{-1} \dfrac{x}{y}$ in degrees (unless specified)

Now let us apply it for the cell where we have generated the complex numbers in its rectangular form.

NOTE: In above explanation we have said that the complex number has a magnitude and direction, so is it okay if we assume and process it as an vector?

```python
w = x_real ** 2 + y_imag ** 2
M = np.sqrt(w)
print(M)
phi = np.arctan2(x_real, y_imag)
print(phi, "In radians")
phi = np.degrees(phi)
print(phi)
```

Now with the above cell we have an Magnitude and angle representation of a complex number, plotting is okay if the user wants to do it.

In our exploration of quadrature signals we will be using the trignometric form of representation of a complex number which is given by $z = x[\cos(\phi) + j\sin(\phi)]$. 

Now we have effectively covered the three major repesentation of complex numbers, one last form exist that we use in math called as the polar form represenatation of the number given by $z = \exp^{(j\phi)}$, this is puzzling but it can be decomposed into the other forms by simple complex transformations.

The expression $z = \exp^{(j\phi)}$ can be expanded as $z = \exp^{(j\phi)} = \cos(\phi) + j\sin(\phi)$
